Phân khúc khách hàng của Trung tâm thương mại

```python
import os
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Tải dữ liệu từ một nguồn online
url = 'https://raw.githubusercontent.com/suhasaitham22/Mall-Customers.csv/refs/heads/main/Mall_Customers.csv'
df = pd.read_csv(url)

# Xem qua dữ liệu
print(df.head())
```

```python
# X chứa các đặc trưng dùng để phân cụm
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
print("\nDữ liệu dùng để phân cụm:")
print(X.head())
```

```python
wcss = []
# Thử các giá trị K từ 1 đến 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # inertia_ chính là giá trị WCSS
```

```python
# Vẽ đồ thị Elbow
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Phương pháp Elbow')
plt.xlabel('Số cụm (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```
